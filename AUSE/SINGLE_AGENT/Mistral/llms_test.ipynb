```python
import pandas as pd

df = pd.read_csv('corpus.csv')
```

```python
import requests

API_URL = "https://api-inference.huggingface.co/models/mistralai/Mixtral-8x7B-Instruct-v0.1"
headers = {"Authorization": f"<HERE YOUR TOKEN>"}
def get_prompt(description_html):
    return (
        "You are an expert app store editor. "
        "Given the following app description in HTML format, summarize it in 2-3 sentences, "
        "with a concise, engaging short description (max 80 characters) suitable for an app store listing. "
        f"App Description HTML:\n{description_html}\n"
        "Format your response as:\n"
        "Short Description: <your short description>\n\n"
        
    )


def query_huggingface_api(prompt):
    payload = {
        "inputs": prompt,
        "parameters": {"max_new_tokens": 128}
    }
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

# Example usage:
from huggingface_hub import HfApi
api = HfApi()
print(api.whoami()) 
```

```python
temp = pd.read_csv('app_prompts_results.csv')
app_names = temp['App Name'].tolist()
print(app_names)
```

```python
results = []
import pandas as pd
import json
try:
    for idx, row in df.iterrows():
        app_name = row['name']
        if app_name in app_names:
            continue
        description_html = row['description_html']
        prompt = get_prompt(description_html)
        try:
            result = query_huggingface_api(prompt)
            print(result)
            
            result_entry = {
                "App Name": app_name,
                "Prompt": prompt,
                "Result": result
            }
            results.append(result_entry)
            # Serialize and append to CSV as DataFrame
            pd.DataFrame([result_entry]).to_csv("app_prompts_results.csv", mode="a", header=not pd.io.common.file_exists("app_prompts_results.csv"), index=False)
            print(result)
        except Exception as e:
            print(f"Error processing app {app_name}: {str(e)}")
except Exception as e:
    error_entry = {
        "App Name": "Global Exception",
        "Prompt": "",
        "Result": f"Global Exception: {str(e)}"
    }
    results.append(error_entry)
    pd.DataFrame([error_entry]).to_csv("app_prompts_results.csv", mode="a", header=not pd.io.common.file_exists("app_prompts_results.csv"), index=False)

results_df = pd.DataFrame(results)
```
